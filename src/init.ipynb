```python
from transformers import AutoTokenizer, AutoModel

model_name = "bert-base-uncased"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

text = "I love machine learning."
inputs = tokenizer(text, return_tensors="pt")

outputs = model(**inputs)
print(outputs.last_hidden_state.shape)  # Output shape: (batch_size, sequence_length, hidden_size)
```

```python
# Import statements
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

# Load the pre-trained model and tokenizer for sequence classification
model_name = "distilbert-base-uncased-finetuned-sst-2-english"

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)


# Tokenize the input text
inputs = tokenizer(
    "Max is desperately trying to find a job.",
    return_tensors="pt"
)

# Run the model and get the logits
with torch.no_grad():
    outputs = model(**inputs)

probs = torch.softmax(outputs.logits, dim=-1)


## Get model information
print(type(outputs))  # Output type: transformers.modeling_outputs.SequenceClassifierOutput
print(outputs)  # Output shape: (batch_size, num_labels)
print(outputs.logits.shape)  # Output shape: (batch_size, num_labels)
print(probs)  # Output probabilities for each class



```
